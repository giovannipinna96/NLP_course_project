### create function for clear text

```python
def clean_text(text):
    '''reduce text to lower-case lexicon entry'''
    lemmas = [token.lemma_ for token in nlp(text) 
              if token.pos_ in {'NOUN', 'VERB', 'ADJ', 'ADV', 'PROPN'}]
    return ' '.join(lemmas)
```

### clear hashtag

```python
#manage #hashtag
import re

hashtag_astrazeneca = []
hashtag_pfizer = []
hashtag_sputnik = []
hashtag_moderna = []

good_hashtag = re.compile('#[A-Za-z0-9\.-_]+')

for tweet in data_astrazeneca['tweet']:
    hashtag_astrazeneca.append(re.findall(good_hashtag, tweet))
data_astrazeneca['hashtag'] = hashtag_astrazeneca

for tweet in data_pfizer['tweet']:
    hashtag_pfizer.append(re.findall(good_hashtag, tweet))
data_pfizer['hashtag'] = hashtag_pfizer

for tweet in data_sputnik['tweet']:
    hashtag_sputnik.append(re.findall(good_hashtag, tweet))
data_sputnik['hashtag'] = hashtag_sputnik

for tweet in data_moderna['tweet']:
    hashtag_moderna.append(re.findall(good_hashtag, tweet))
data_moderna['hashtag'] = hashtag_moderna
```

```python
#manage #hashtag
import re

good_hashtag = re.compile('#[A-Za-z0-9\.-_]+')

def clean_hashtag(text):
  '''replace all hashtag with hash'''
  twt_encode = re.sub(good_hashtag, ' ', text)
  return  twt_encode
```

```python
# manage link
import re

site = re.compile('http[A-Za-z0-9\.-_/:]+')

def clean_site(text):
  '''replace all sitelink with wwwsite'''
  twt_encode = re.sub(site, ' ', text)
  return twt_encode
```

### clear special caracters

```python
#manage speical caracters
import re

speical_caracters = re.compile('\W')

def clean_special_caracters(text):
  '''replace all speical caracters with space'''
  twt_encode = re.sub(speical_caracters, ' ', text)
  return twt_encode
```

### clean digits

```python
#manage numbers
import re

digits = re.compile('[0-9]+[.,]*[0-9]*')

def clean_digits(text):
  '''replace all digits with 0'''
  twt_encode = re.sub(digits, '0', text)
  return twt_encode
```

### clear location

```python
#manage cites
import re

def clear_roma(text):
    tx = re.sub('Rom[A-Za-z][,]?[A-Za-z]*','Rome',text)
    return tx

def clear_milano(text):
    tx = re.sub('Milan[A-Za-z]+[,]?[A-Za-z]*','Milano',text)
    return tx


def clear_napoli(text):
    tx = re.sub('Nap[A-Za-z]+[,]?[A-Za-z]*','Napoli',text)
    return tx


def clear_italia(text):
    tx = re.sub('Ital[Y-yA-Za-z\W\w]+[,]?[A-Za-z]*','Italia',text)
    return tx


def clear_torino(text):
    tx = re.sub('Torin[A-Za-z]+[,]?[A-Za-z]*','Torino',text)
    return tx


def clear_other(text):
    tx = re.sub('[\W]*','',text)
    return tx

def clear_lombardia(text):
    tx = re.sub('Lombard[A-Za-z]+[,]?[A-Za-z]*','Lombardia',text)
    return tx
```

```python
def clear_location(text):
    text = clear_other(text)
    text = clear_roma(text)
    text = clear_milano(text)
    text = clear_napoli(text)
    text = clear_italia(text)
    text = clear_torino(text)
    text = clear_lombardia(text)
    return text
```

## Checklist: how to classify my data

1. label at ***least 2000*** tweets in your data set as `positive`or `negative`
2. preprocess the text of *all* tweets in your data (labeled and unlabeled)
3. read in the labeled tweets and their labels
4. transform the labels into numbers
5. use `TfidfVectorizer` to extract the features and transform them into feature vectors
6. select the top $N$ features (where $N$ is smaller than the number of labeled tweets)
7. create a classifier
8. use 5-fold CV to find the best regularization parameter, top $N$ feature selection, and maybe feature generation and preprocessing steps

Once you are satisfied with the results:
9. read in the rest of the (unlabeled) tweets
10. use the `TfidfVectorizer` from 5. to transform the new data into vectors
11. use the `SelectKBest` selector from 6. to get the top $N$ features
12. use the classifier from 7. to predict the labels for the new data
13. save the predicted labels or probabilities to your database or an Excel file

```python
#replace words by their numerical IDs and their frequency
#dobbiamo trasformare il corpus con tf-idf
print("translating corpus to IDs", flush=True)
ldacorpus = [dictionary.doc2bow(text) for text in instances] #lo chiamiamo per ogni text in istances
# learn TFIDF values from corpus
print("tf-idf transformation", flush=True)
tfidfmodel = TfidfModel(ldacorpus)
# transform raw frequencies into TFIDF
model_corpus = tfidfmodel[ldacorpus]
```

```python
coherence_values = []

for num_topics in range(5, 20): 
    model = LdaMulticore(corpus=model_corpus,  
                         id2word=dictionary, 
                         num_topics=num_topics)

    coherencemodel_umass = CoherenceModel(model=model, 
                                          texts=instances, 
                                          dictionary=dictionary, 
                                          coherence='u_mass')

    coherencemodel_cv = CoherenceModel(model=model, 
                                       texts=instances, 
                                       dictionary=dictionary, 
                                       coherence='c_v')

    umass_score = coherencemodel_umass.get_coherence()
    cv_score = coherencemodel_cv.get_coherence()
    
    print(num_topics, umass_score, cv_score)
    coherence_values.append((num_topics, umass_score, cv_score))
```

```python
limit=50000

data_tot['clean_text'] = data_tot.clean_text.apply(str)

instances = data_tot.clean_text.apply(str.split)[:limit] 
print("creating dictionary", flush=True)

dictionary = Dictionary(instances)


len(dictionary.keys()) 
```

```python
print("translating corpus to IDs", flush=True)
ldacorpus = [dictionary.doc2bow(text) for text in instances] 

print("tf-idf transformation", flush=True)
tfidfmodel = TfidfModel(ldacorpus)

model_corpus = tfidfmodel[ldacorpus]
```

```python

i = 10
print(data_tot.tweet.iloc[i])
print(instances[i])
print(ldacorpus[i]) 
print(model_corpus[i])
```

```python
coherence_values = []

dev_size = 1000000
eval_size = 5000

for num_topics in range(5, 20):
    model = LdaMulticore(corpus=model_corpus,  
                         id2word=dictionary, 
                         num_topics=num_topics)

    coherencemodel_umass = CoherenceModel(model=model, 
                                          texts=instances, 
                                          dictionary=dictionary, 
                                          coherence='u_mass')

    coherencemodel_cv = CoherenceModel(model=model, 
                                       texts=instances, 
                                       dictionary=dictionary, 
                                       coherence='c_v')

    umass_score = coherencemodel_umass.get_coherence()
    cv_score = coherencemodel_cv.get_coherence()
    
    print(num_topics, umass_score, cv_score)
    coherence_values.append((num_topics, umass_score, cv_score))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
seaborn.set_context('poster') 


scores = pd.DataFrame(coherence_values, columns=['num_topics', 'UMass', 'CV'])
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
scores.plot.line(x='num_topics', y='UMass', ax=ax[0], xticks=range(5,20));
scores.plot.line(x='num_topics', y='CV', ax=ax[1], xticks=range(5,20));
```

```python
num_topics = 6

num_passes = 10
chunk_size = len(model_corpus) * num_passes/200
print(chunk_size)

start = time.time()
print("fitting model", flush=True)
model = LdaMulticore(num_topics=num_topics, # number of topics
                     corpus=model_corpus, # what to train on 
                     id2word=dictionary, # mapping from IDs to words
                     workers=min(10, multiprocessing.cpu_count()-1), # choose 10 cores, or whatever computer has
                     passes=num_passes, # make this many passes over data
                     chunksize=chunk_size, # update after this many instances
                     alpha=0.5
                    )
    
print("done in {}".format(time.time()-start), flush=True)
```

```python
import re


topic_sep = re.compile(r"0\.[0-9]{3}\*") 
model_topics = [(topic_no, re.sub(topic_sep, '', model_topic).split(' + ')) for topic_no, model_topic in
                model.print_topics(num_topics=num_topics, num_words=10)]

descriptors = []
for i, m in model_topics:
    print(i+1, ", ".join(m[:10]))
    descriptors.append(", ".join(m[:2]).replace('"', ''))
  
```
